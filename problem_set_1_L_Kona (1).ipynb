```python
## Histogram

Driver_Age = stops['Driver_Age']
n_data = len(Driver_Age)

# Number of bins is the square root of number of data points: n_bins
n_bins = np.sqrt(n_data)
n_bins = int(n_bins)
print(f"Number of bins: {n_bins}")


# Plot the histogram
fig = plt.figure(figsize=(12,6))

_ = plt.hist(Driver_Age, bins=n_bins, color='green')
_ = plt.xlabel('Driver Age', labelpad=20)
_ = plt.ylabel('count', labelpad=20)
_ = plt.title("Bins computed using the Square root rule")

```

- Histograms with large bin widths that group data into only a few bins. This does not give good information about variability in the distribution.

- Histograms with small bin widths that group data into lots of bins. This does not help us see the pattern in the data.


Impact of more / less no. of bins doesn't represent the data well. 

**bins='auto' :: **Matplotlib determines bins using SturgesÂ andÂ Freedman-DiaconisÂ bin choice methods. 

So, histogram with 20-30 bins would give appropriate pattern for this data.

#### 9. Plot a line chart with the number of traffic stops for each month (hint: think about the aggregations we did in class).

```python
## Create a dataframe named stops_count for this graph that has a column named Month_of_Stop 
## Also create a column named count representing the ammount of stops per month  https://sparkbyexamples.com/pandas/pandas-groupby-count-examples/

stops_count = stops['Month_of_Stop'].value_counts().rename_axis('Month_of_Stop').reset_index(name='counts')
stops_count = stops_count.sort_values(by="Month_of_Stop", ascending=True)
stops_count
```

#### 11. Plot the same plot but add in `.facet()` by the `Reason_for_Stop` variable.

```python
## create the dataframe that groups by month and reason for stop 
stops_facet = stops.groupby(['Month_of_Stop','Reason_for_Stop'])['ObjectID'].count().reset_index(name='counts')

stops_facet = stops_facet.sort_values(by="Month_of_Stop", ascending=True)
stops_facet
```

```python
## Facet chart https://stackoverflow.com/questions/64770801/python-altair-facet-line-plot-with-multiple-variables

alt.Chart(stops_facet, title="Stops per Month by 'Reason_for_Stop' facet variable").mark_line().encode(
    x='Month_of_Stop:T',
    y='counts:Q',
    facet=alt.Facet('Reason_for_Stop:O').columns(3)
)
```

#### 12. What is a problem with this plot? xxxxx

Issue::

A Facet chart is used to visualize a distribution and compare among multiple variables/dimensions. 

All charts using the same X and Y axis. Cannot display their axis or scale independently.



Solution::

Creating a facet_wrap(), we can style and scale each subplot as per its measure values independently.



To address this problem, you will need to figure out how to adjust the scale. To do this, you need to use altairs documentation to see whether there is a function that could help you.

https://altair-viz.github.io/user_guide/scale_resolve.html

#### 13. What parameter allows you to modify the scales of `facet_wrap`? 

**Ans::** 'Chart.resolve_scale()'


For facet_wrap(), you can control whether the position scales are the same in all panels (fixed) or allowed to vary between panels (free) with the resolve_scale parameter


#### 14. Plot the same plot but with a free y-axis scale. (This may take some research but very findable)

```python
# Updated Facet Chart
alt.Chart(stops_facet).mark_line().encode(
    x='Month_of_Stop:T',
    y='counts:Q',
    facet=alt.Facet('Reason_for_Stop:O').columns(3)
).resolve_scale(
    y='independent'
)

```

**Ans::**

If the scales = "free_y" argument is added, their height will be proportional to the length of the y scale;

Similarly, if "free_x" their width will be proportional to the length of the x scale; or if "free" both height and width will vary. 

Using facet_wrap, each plot is displayed independently, and it can "free" its x-axis and y-axis.
